```python
import torch
import torch.nn.functional as F
x1 = torch.tensor(6.7)
y1 = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
z1 = w*x1 + b
```

```python
y_pred = torch.sigmoid(z1)
y_pred
```

```python
loss = F.binary_cross_entropy(y_pred,y1)
loss
```

```python
print(w.grad)
print(b.grad)
```
